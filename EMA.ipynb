```python
import pandas as pd
import datetime
from google.colab import drive
drive.mount('/content/drive')
```

```python
basepath = "/content/drive/MyDrive/ISEF/Datasets/MergedPriceDataCalculatedColumnscsv.csv"
dfbase = pd.read_csv(basepath)
dfEMA = pd.DataFrame()
dfEMA["Date"] = dfbase["Date"]
for i in range(182):
  columnname = dfbase.columns[i+2]
  dfEMA[columnname + "-EMA4"] = dfbase[columnname].ewm(span=4, adjust=False).mean()
```
